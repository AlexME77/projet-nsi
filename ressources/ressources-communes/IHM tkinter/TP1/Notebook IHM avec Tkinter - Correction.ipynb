#	Créer une fenêtre 


```python
# Importer les classes du module tkinter
from tkinter import *
# Création  d'une fenêtre
fenetre = Tk()
# Titre de la fenêtre
fenetre.title("Ma première fenêtre")
# Dimension de la fenêtre 400x300 pixels
fenetre.geometry("400x300")
# Lancement du gestionnaire d'événements
fenetre.mainloop()
```

## Ce qu'il faut retenir :

•	<b>fenetre=Tk()</b> crée un objet de type fenêtre qui s'appellera «fenetre»
Une fenêtre est un objet de type surface. Cet objet est créé par la classe <b>Tk</b>. 
La méthode geometry() de la classe Tk permet de fixer les dimensions de la fenêtre selon les coordonnées (x, y).
<img src="taille_fenetre.jpg">

•	fenetre.mainloop() c'est la ligne qui provoque le démarrage du réceptionnaire d'événements associé à la fenêtre. Cette instruction est nécessaire pour que l'application soit à l'affût des événements appliqués sur l'objet «fenetre» comme clic de souris, appui sur une touche du clavier, etc.

Remarque : la fenêtre s'ajuste automatiquement au contenu si on ne précise pas ses dimensions.
Pour détruire la fenetre, on peut appeler la fonction <b>fenetre.destroy()</b>

# 5	Classe  «Entry»

La classe «Entry» est utilisée pour définir les widgets permettant de saisir des données. Il faut donc prévoir une variable permettant de récupérer les données saisies. Pour cela, on utilise une variable de contrôle.
Une variable de contrôle est un objet qui se comporte comme une variable ordinaire de Python dans le sens que c'est un conteneur pour une valeur comme une chaine de caractères ou un nombre. L'intérêt des variables de contrôle, c'est qu'elles sont partagées entre plusieurs widgets. Cela signifie que si le contenu d'une variable de contrôle est modifié, tous les widgets qui sont reliés à cette variable de contrôle sont automatiquement mis à jour sur l'écran.

Pour obtenir une variable de contrôle, utiliser l'un des constructeurs, en fonction du type de valeur qu'on souhaite mémoriser :

•	valeur = DoubleVar()	contient un nombre à virgule flottant ; sa valeur par défaut est 0.0,

•	valeur = IntVar()	contient un nombre entier ; sa valeur par défaut est 0,

•	valeur = StringVar()	contient une chaine de caractères ; sa valeur par défaut est "".

Toutes les variables de contrôle disposent des méthodes : 

•	get() 	retourne la valeur courante de la variable de contrôle, 

•	set(valeur) 	modifie la valeur de la variable de contrôle.

Compléter les lignes de commentaires du script ci-dessous puis l'exécuter:


```python
from tkinter import *
fenetre=Tk()
#On définit un texte affichant "Cliquer et saisir", et on lui précise une taille et une couleur(noir)
texte=Label(fenetre, text='Cliquer et saisir :', width=20, height=3, fg="black")
texte.pack()
#On définit une variable de contrôle composée d'une chaine vide, puis on affiche cette chaine, en rouge sur fond blanc
saisie=StringVar()
label = Label(fenetre,textvariable = saisie,)
label.pack()
#On définit une classe entry afin de pouvoir saisir des données, sous forme de chaîne de caractères
entree=Entry(fenetre,textvariable=saisie, width=30)
entree.pack()
#On définit un bouton permettant de quitter, en "détruisant" la fenêtre lorsqu'on appuie dessus
bouton=Button(fenetre , text='Quitter' , command=fenetre.destroy)
bouton.pack()
fenetre.mainloop()
```

# 6 Exercice

Écrire un script qui demande un nombre compris entre 1-100 et qui affiche «Gagné» si le nombre entrée est la valeur du nombre tiré au hasard par le programme, «perdu» sinon. Le contenu doit s'effacer pour pouvoir rejouer. Pour cela créer la fonction vérifier() qui sera exécutée dès que le bouton «Valider» est pressé. Cette fonction récupère le contenu de la saisie et le compare avec le nombre tiré au hasard.

<img src="exercice6.jpg">

```python
#Correction exercice 6


#Importation des modules nécessaires.
from tkinter import *
from random import *


fenetre=Tk()
fenetre.title('Jeu deviner un nombre')
fenetre.geometry('300x150') # définition des dimensions de la fenêtre (300x150)

# définition de la zone de saisie
saisie=StringVar()
entree=Entry(fenetre,textvariable=saisie,width=15)
entree.pack(side=TOP)


def lireValeur():
    saisie = entree.get()
    return int(saisie)

nb=randint(1,100)

def verifier():
    if nb == lireValeur():
        texte.config(text="Gagné!!!:D")
        texte.pack()
    elif nb > lireValeur():  
        texte.config(text="la valeur à trouver est supérieure")
        texte.pack()        
    else:
        texte.config(text="la valeur à trouver est inférieure")
        texte.pack()   
texte=Label(fenetre, text='Veuillez proposer un nombre compris entre 1-100',width=60,height=3,fg="black")
texte.pack()

#On définit  les boutons valider et quitter.

bouton=Button(fenetre,text="Quitter",command=fenetre.destroy)
bouton.pack(side=BOTTOM)
bouton=Button(fenetre,text="Valider",command=verifier)
bouton.pack(side=BOTTOM)
fenetre.mainloop()
```

# 8	Taille et positionnement des fenêtres sur l’écran

La taille et la position de la fenêtre est définie à l’aide de la méthode geometry().
Exécuter sans paramètre, elle renvoie la taille et la position de la fenêtre sous la forme d’une chaine de caractères composée da la façon suivante :

"LxH+X+Y"
•	L	largeur de la fenêtre,

•	x	séparateur,

•	H	hauteur de la fenêtre,

•	+	séparateur (en fait + ou -),

•	X	position du côté gauche de la fenêtre par rapport au côté gauche de l’écran,

•	+	séparateur (en fait + ou -),

•	Y	position du côté haut de la fenêtre par rapport au côté haut de l’écran.

Exemple : "50x100+45+90"  => L = 50, H=100, X = 45, Y = 90
