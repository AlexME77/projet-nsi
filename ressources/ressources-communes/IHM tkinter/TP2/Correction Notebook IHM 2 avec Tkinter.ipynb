# 2	Canevas : Canvas

Le canevas représente la surface pour dessiner. En fait, c'est un conteneur de dessin.
Pour créer un conteneur de dessin, on utilise la ligne de script suivante :

<b>canevas = Canvas(fenetre, bg = 'black', height = 200, width = 200) 
    
canevas.pack()</b>
### Ce qu'il faut retenir :
•	«canevas» est le nom d'objet du conteneur de dessin de la classe Canvas,

•	«canevas.pack()»  la méthode «pack()» va ajuster la taille de l'objet «fenetre» à la taille du canevas,

•	bg = 'black' le paramètre «bg» donne la couleur de fond du canevas (bd est background en anglais).

De nombreuse couleurs sont disponibles : red, blue, yellow, white, green, gray, pink, orange, dark red, magenta, turquoise, gold,...

Le canevas est muni d'un repère :

•	le point en haut à gauche de la fenêtre a pour coordonnées (0,0);

•	le point en bas à droite, par exemple d'un canevas placé dans une fenêtre de 200x200, sera le point coordonnée (200,200). 

•	le point central a pour coordonnée (largeur de fenêtre /2, hauteur de la fenêtre / 2).

## 2-1	Dessiner une ligne

Pour dessiner une ligne, on utilise la méthode create_line() de la classe Canvas. Le code peut être : 

<b>Canvas.create_ligne(x1, y1, x2, y2,  paramètres...)</b>

Cela permet de tracer une ligne dans le canevas du point (x1, y1) au point (x2, y2)

•	fill	pour préciser la couleur de la ligne,

•	width	pour préciser la largeur de la ligne.

### Exécuter le script suivant pour dessiner une ligne.


```python
from tkinter import *
fenetre=Tk()
fenetre.title('Canevas')
canevas = Canvas(fenetre,bg='white', width=200, height=100)
canevas.create_line(10,50,190,50, width=5, fill='gray')
canevas.pack()
fenetre.mainloop()
```

<b><u>Remarque :</u></b> On peut dessiner plusieurs lignes en une seule ligne de script en donnant d'autres points de coordonnées (x,y) :
<b>canevas.create_line(x1,y1,x2,y2,x3,y3)</b>

### Exécuter le script affichant plusieurs figures géométriques :

# 7	Complément sur les événements pour répondre aux actions de l'utilisateur
Un événement est la survenu de quelque chose dans l'application. Par exemple, l'utilisateur appuie sur une touche du clavier, clique ou déplace la souris. L'application à certainement besoin de réagir à ces actions de l'utilisateur.
Les widgets ont des comportements prédéfinis. Par exemple un bouton réagira à un clic de souris en appelant la fonction associée à son argument «command».
Cependant, tkinter fourni les moyens pour ajouter, changer ou supprimer de tels comportement.
## 7-1	Définitions
•	Un événement (event) est survenue d'une action (clavier, souris),
•	un gestionnaire d'événement (event handler) est une fonction de l'application qui a vocation a être appelée lorsqu'un n événement se produira,
•	On parle de liaison lorsque l'application défini un gestionnaire d'événement qui sera effectivement appelé lorsqu'un certain événement se produit sur widget.
## 7-2	Niveau de liaison
On peut lier un gestionnaire à un événement à l'un de ces trois niveaux :
•	Liaison au niveau d'un widget.
Pour lier un événement à un widget, on appelle la méthode «bind()» sur ce widget. Par exemple, si on a un canevas et qu'on souhaite dessiner un disque orange sur celui-ci à chaque fois que l'utilisateur appuie sur le bouton 2 (celui du milieu) de la souris. Pour définir ce comportement :
canevas.bind('<Button-2>', dessinerDisqueOrange)
Le premier argument est un descripteur de séquence d'événement qui indique à tkinter que lorsque le bouton central est appuyé, il faut appeler le gestionnaire d'événement dessinerDisqueOrange fourni comme deuxième argument,
•	Liaison au niveau d'une classe.
Pour lier un événement à tous les widgets d'une classe, on appelle la méthode «bind_class()».
Par exemple, si on a plusieurs canevas, si on veut régler les choses de telle sorte que lorsque l'utilisateur clique sur le bouton gauche le la souris, un disque orange soit tracé dans celui sur le quel se trouve la souris. Plutôt d'appeler la méthode «bind()» pour chaque canevas.
#canevas est un widget arbitraire
canevas.bind_class('Canvas', '<Button-2>', dessinerDisqueOrange)
•	Liaison au niveau de l'application.
On peut définir une liaison d'événement de telle sorte que le gestionnaire d'événement soit appelé indépendamment du widget qui a le focus ou qui se trouve sous la souris.
Par exemple, si on souhaite lier un événement «appui sur la touche ImprÉcran» à tous les widgets de l'application de telle sorte que l'écran soit imprimé indépendamment du widget qui a effectivement reçu l'appui sur la touche. Pour lier un événement au niveau de l'application, appeler la méthode «bind_all()». 
#canevas est un widget arbitraire
canevas.bind_all( '<Key-Print>', imprimerEcran)
## 7-3	Séquence d'événements
Une séquence d'événements est une chaine de caractères contenant un ou plusieurs motifs d'événements.
•	le motif est enfermé entre des chevrons <...>,
•	le type d'événement décrit le genre de celui-ci, comme appui sur une touche «KeyPress», ou un clic de souris,
•	on peut indiquer un ou plusieurs modificateurs avant son type pour décrire une combinaison comme un appui sur la touche «Maj» ou «Control» pendant qu'une autre touche ou qu'un bouton de la souris est enfoncé.
Quelques motifs d'événements
•	<Button-1>	l'utilisateur a appuyé sur le premier bouton  (bouton gauche) de la souris,
•	<Double-1>	l’utilisateur a double-cliqué sur le bouton gauche de la souris,
•	<Enter>	l’utilisateur a appuyé sur la touche Entrée
•	<KeyRelease-H>	l'utilisateur a relâché la touche H,
•	<Control-Shift-KeyPress-H>	l'utilisateur a appuyé simultanément sur les touches Control, Maj et H.

## 7-4 Evènemenents Clavier

L' application Tkinter srutent en boucle les événements (events). Ces événements peuvent provenir du clavier (touche frappée, enfoncée, relâchée), de la souris (clic gauche ou droit, glisser/déplacer...) ou du programme lui-même.

La détection de ces événements est automatique, il suffit de spécifier comment lier un événement particulier à l'action qui doit lui correspondre. C'est l'objet de la méthode bind() (lier en anglais).
Sa syntaxe est de la forme cible.bind(événement e, callback) où la fonction associée callback admet au minimum un paramètre formel noté, pour des raisons de clarté, event ou e (mais tout autre nom de variable convient). Cette ligne de code se lit : "au cas où l'événement e se produit, appeler la fonction callback(e)".
cible désigne une fenêtre, un cadre, un widget ou un objet graphique.
L'attribut event.char (e.char) contient alors le caractère tapé.

Il est possible de surveiller plusieurs événements simultanément en utilisant plusieurs fois la méthode bind().

Nom des événements liés aux touches du clavier :

    "<Key>" désigne une touche quelconque ;
    "A", "a", "1" désignent respectivement « A », « a » et « 1 » ;
    "<Up>", "<Down>", "<Left>", "<Right>", désignent les flèches de direction ↑ ↓ ← → ;
    "<Return>", "<Cancel>", "<F1>", "<Space>"…

Exemple :  fenetre.bind("<Left>", lambda e: deplacerGauche(e)) appelle la fonction deplacerGauche(event) lorsque la flèche gauche ← du clavier est frappée. Attenion, la fonction deplacerGauche(e) devra prendre un argument un objet de type event.
